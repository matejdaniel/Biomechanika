# Stupně volnosti

> **Stupeň volnosti** (DOF, degrees of freedom) je v mechanice počet nezávislých parametrů, které definují konfiguraci mechanického systému. Jinými slovy, je to počet nezávislých pohybů, které může těleso vykonávat.

* V ploše má bod dva stupně volnosti 
    * posun podél osy x a podél osy y
* V ploše má těleso pak tři stupně volnosti 
    * posun podél osy x, osy y
    * otočení kolem bodu.
* V prostoru má bod tři stupně volnosti
    * posuny podél os x, y, z 
* V trojrozměrném prostoru má tuhé těleso 6 stupňů volnosti:
    * 3 posuvné (pohyb podél os x, y, z)
    * 3 rotační (rotace kolem os x, y, z)

> **Vazba** je omezení, které omezuje nebo zakazuje některé stupně volnosti tělesa. Vazby jsou realizovány různými strukturami, jako jsou klouby, vazy, svaly a další tkáně.

Každá vazba **odebírá** tělesu jeden nebo více stupňů volnosti. Například loketní kloub odebírá 5 stupňů volnosti a umožňuje pouze flexi a extenzi v jedné rovině.

## Klouby

![klouby](https://assets.puzzlefactory.com/puzzle/275/058/original.webp)

### 1. Kloub kulový (articulatio spheroidea)

Vyznačuje se tím, že kulovitá hlavice jedné kosti zapadá do kulovité jamky na druhé kosti. Jedná se o nejpohyblivější tvar kloubu, který umožňuje pohyby ve všech směrech

* **Stupně volnosti:** 3 (pohyb ve všech třech rovinách)
    * Flexe/extenze (ohýbání/natahování)
    * Abdukce/addukce (odtahování/přitahování)
    * Rotace (vnitřní/zevní)
* **Příklady:**
    * **Ramenní kloub:** Umožňuje širokou škálu pohybů paže, jako je zvedání, kroužení a otáčení. 
    * **Kyčelní kloub:** Umožňuje pohyb nohy ve všech směrech, což je důležité pro chůzi, běh a další aktivity. 
### 2. Kloub čepový (articulatio trochoidea)

Okrouhlý konec jedné kosti zapadá do kroužku, který je tvořen druhou kostí a obklopujícím vazem. Tento typ kloubu umožňuje rotaci v podélné ose kosti.

* **Stupně volnosti:** 1 (rotace kolem jedné osy)
* **Příklady:**
    * **Spojení mezi prvním a druhým krčním obratlem (atlas a axis):** Umožňuje otáčení hlavy do stran. V biomechanice se zkoumá například při analýze otáčení hlavy při řízení auta.

### 3. Kloub šarnýrový  (articulatio ginglymus)

Umožňuje pohyb kloubu pouze ve dvou směrech (např. dopředu a dozadu) a vyznačuje se tím, že válcovitý konec jedné kosti zapadá do žlábku na povrchu druhé kosti.

* **Stupně volnosti:** 1 (flexe a extenze v jedné rovině)
* **Příklady:**
    * **Loketní kloub:** Umožňuje ohýbání a natahování paže. V biomechanice se zkoumá například při analýze hodu míčem nebo zvedání břemen.

### 4. Kloub plochý (articulatio plana)

Vyznačuje se tím, že jeho kloubní povrchy jsou v podstatě ploché destičky, které umožňují pouze rotaci a krátké klouzavé pohyby. 

* **Stupně volnosti:** Omezený pohyb v různých směrech (klouzání po sobě)
* **Příklady:**
    * **Klouby mezi obratli páteře:** Umožňují drobné pohyby páteře, které se skládají do celkového pohybu. V biomechanice se zkoumá například při analýze držení těla a pohybu páteře při chůzi.
    * **Klouby mezi zápěstními a zánártními kostmi:** Umožňují drobné pohyby ruky a nohy, které se podílejí na celkové hybnosti.

### 5. Kloub vejčitý (articulatio ellipsoidea)

* **Stupně volnosti:** 2 (flexe/extenze a abdukce/addukce)
* **Příklady:**
    * **Zápěstní kloub:** Umožňuje ohýbání a natahování ruky a pohyb do stran. V biomechanice se zkoumá například při analýze úchopu předmětů a psaní.
    * **Hlezenní kloub:** Umožňuje ohýbání a natahování nohy a pohyb do stran (everze a inverze). V biomechanice se zkoumá například při analýze chůze, běhu a skoku.

### 6. Kloub sedlovitý (articulatio sellaris)

* **Stupně volnosti:** 2 (flexe/extenze a abdukce/addukce) + omezená rotace
* **Příklady:**
    * **Karpometakarpální kloub palce:** Umožňuje palci pohyb v širokém rozsahu, což je důležité pro uchopování a manipulaci s předměty. V biomechanice se zkoumá například při analýze úchopu a jemné motoriky.

## Pohyby v kloubech

Popisujeme tři základní pohyby mezi styčnými plochami kloubu:
* rotace
* pohyb valivý
* pohyb smykový

Dále pohyb závisí na počtu os, ve kterých se pohyb děje, a na počtu kostí, které jsou součástí kloubu.
* pohyb okolo osy X (rovina frontální) – **abdukce, addukce**
* rotační pohyb okolo osy Y (rovina sagitální) – **flexe, extenze**
* pohyb okolo osy Z (rovina transverzální) – **vnitřní, zevní rotace**

kombinací těchto pohybů vzniká cirkumdukce

## Druhy kloubů dle počtu os a tvaru styčných ploch

Rotační klouby můžeme dělit na:
* **Klouby tříosé** - Pohyb se děje ve všech třech osách. Příkladem takového kloubu je kloub kulovitý.
* **Klouby dvojoosé** - Pohyb se děje podle os flexe-extenze a abdukce-addukce. Kloub elipsovitý a kloub sedlovitý.
* **Klouby jednoosé** - Obecně se oznčuje jako kloub válcový, v anatomiii se využívá označení:
    * kloub šarnýrový (ginglymus)- osa pohybu kolmá na podélnou osu kosti
    * kloub kolový – osa otáčení shodná s podélnou osou kosti (radioulnární kloub)

Mezi rotační klouby nepatří plochý kloub, který umožňuje jenom translaci.


## Osteokinematika a artrokinematika

> **Osteokinematika** popisuje zřetelné pohyby kostí, které jsou viditelné zvenčí. Jsou to hrubé pohyby, které se dějí mezi dvěma kostmi. Vznikají rotací kolem osy kloubu.

Některé příklady jsou:
* Flexe a extenze (ohýbání a natahování)
* Abdukce a addukce (odtažení a přitažení)
* Laterální (zevní) rotace a mediální (vnitřní) rotace

Osteokinematika se liší od artrokinematiky.  

> **Artrokinematika** popisuje malé pohyby, které se dějí na samotné kloubní ploše. Pohyby jsou valivé, klouzavé/posuvné a rotační.
