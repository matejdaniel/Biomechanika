## **Initial Contact (Počáteční kontakt nebo Heel Strike)**

| Kloub      | Kinematika                                                                     | Svalová aktivita                                                                                              |
|------------|--------------------------------------------------------------------------------|-----------------------------------------------------------------------------------------------------------------|
| Pánve      | Pánve je v neutrální pozici, připravená přenést váhu na jednu nohu              | Aktivace stabilizátorů pánve, zejména [Gluteus medius](Gluteus_medius) pro podporu stability               |
| Kyčel      | Kyčel je ve střední pozici, mírně flexován, připraven na kontakt s podložkou   | Aktivace [Iliopsoas](Iliopsoas) pro pokračující flexi kyčle a stabilizaci během kontaktu s podložkou      |
| Koleno     | Koleno je v plné extenzi, připraveno absorbovat náraz při kontaktu s podložkou | Aktivace extenzorů kolene, především [Quadriceps](Quadriceps), pro ochranu kloubu a přenos síly do další fáze |
| Ankle (kotník) | Kotník je v pozici dorsiflexe, připraven absorbovat náraz při kontaktu s podložkou | Aktivace [Tibialis anterior](Tibialis_anterior) pro kontrolu nad dorsiflexí a přípravu na stabilizaci pod patou |


## **Foot Flat (Úplný kontakt s podložkou)**

| Kloub      | Kinematika                                                                     | Svalová aktivita                                                                                              |
|------------|--------------------------------------------------------------------------------|-----------------------------------------------------------------------------------------------------------------|
| Pánve      | Pánve pokračuje v rotaci směrem k opačné straně, stabilizuje celkové nastavení | Aktivace [Gluteus medius](Gluteus_medius) a [Gluteus maximus](Gluteus_maximus) pro stabilizaci během přechodu do fáze plného kontaktu |
| Kyčel      | Kyčel se nachází v mírné flexi, připravuje se na přenos síly z nohy do těla    | Aktivace [Rectus femoris](Rectus_femoris) a [Iliopsoas](Iliopsoas) pro pokračující flexi kyčle             |
| Koleno     | Koleno se nachází v mírné flexi, připraveno absorbovat sílu přenášenou na nohu | Aktivace [Quadriceps](Quadriceps) a [Hamstrings](Hamstrings) pro kontrolu nad flexí a extenzí kolene       |
| Ankle (kotník) | Kotník se nachází ve střední pozici mezi dorsiflexí a plantární flexí       | Aktivace [Gastrocnemius](Gastrocnemius) a [Soleus](Soleus) pro stabilizaci kotníku a podporu tělesné hmotnosti |


## **Mid-Stance (Střední fáze)**

| Kloub      | Kinematika                                                                     | Svalová aktivita                                                                                              |
|------------|--------------------------------------------------------------------------------|-----------------------------------------------------------------------------------------------------------------|
| Pánve      | Pánve je stabilizována a udržuje rovnováhu těla, pokračuje v mírné rotaci       | Aktivace [Gluteus medius](Gluteus_medius) pro kontrolu nad stabilitou pánve a udržení rovnováhy               |
| Kyčel      | Kyčel je v mírné extenzi, přenáší tělesnou hmotnost z jedné nohy na druhou      | Aktivace [Gluteus maximus](Gluteus_maximus) a [Iliopsoas](Iliopsoas) pro zajištění stability a kontroly nad přenosem hmotnosti |
| Koleno     | Koleno je mírně ohnuto, absorbující sílu přenášenou z tělesné hmotnosti          | Aktivace [Quadriceps](Quadriceps) pro kontrolu nad extenzí kolene a stabilizaci během přechodu do fáze odrazu |
| Ankle (kotník) | Kotník je v pozici mírné plantární flexe, připravený přenášet sílu            | Aktivace [Gastrocnemius](Gastrocnemius) a [Soleus](Soleus) pro stabilizaci a kontrolu nad přenosem síly       |

## **Heel-Off (Odrazová fáze)**

| Kloub      | Kinematika                                                                     | Svalová aktivita                                                                                              |
|------------|--------------------------------------------------------------------------------|-----------------------------------------------------------------------------------------------------------------|
| Pánve      | Pánve pokračuje v rotaci, připravuje se na odraz nohy                         | Aktivace [Gluteus maximus](Gluteus_maximus) pro kontrolu nad stabilizací pánve při přechodu do odrazové fáze  |
| Kyčel      | Kyčel je ve větší flexi, připraven se odrazit a posunout tělesnou hmotnost dopředu | Aktivace [Iliopsoas](Iliopsoas) pro zajištění správné flexe kyčle a podporu odrazu                           |
| Koleno     | Koleno je mírně ohnuto, připraveno pro odraz                                  | Aktivace [Hamstrings](Hamstrings) pro zajištění extenze a kontrolu pohybu kolene při odrazu                  |
| Ankle (kotník) | Kotník je v plantární flexi, připravený pro odraz                             | Aktivace [Gastrocnemius](Gastrocnemius) a [Soleus](Soleus) pro maximální stabilizaci kotníku při odrazu        |

## **Toe-Off (Odrazový bod)**

| Kloub      | Kinematika                                                                     | Svalová aktivita                                                                                              |
|------------|--------------------------------------------------------------------------------|-----------------------------------------------------------------------------------------------------------------|
| Pánve      | Pánve pokračuje ve stabilizaci a rotaci na opačnou stranu                      | Aktivace [Gluteus medius](Gluteus_medius) pro stabilizaci pánve a její rotaci směrem dopředu                  |
| Kyčel      | Kyčel dosahuje maximální flexe, připravený na následný krok                    | Aktivace [Iliopsoas](Iliopsoas) pro udržení flexe a přípravu na posun tělesné hmotnosti                      |
| Koleno     | Koleno se nachází v maximálním ohnutí, připraveno pro uvolnění a přenos síly    | Aktivace [Hamstrings](Hamstrings) pro zajištění správného ohnutí a připravení na přenos síly do další fáze    |
| Ankle (kotník) | Kotník se nachází v maximální plantární flexi, připravený pro odraz         | Aktivace [Gastrocnemius](Gastrocnemius) a [Soleus](Soleus) pro maximální stabilizaci kotníku a generování síly k dalšímu kroku |

## **Swing Phase (Fáze kyvného pohybu)**

| Kloub      | Kinematika                                                                     | Svalová aktivita                                                                                              |
|------------|--------------------------------------------------------------------------------|-----------------------------------------------------------------------------------------------------------------|
| Pánve      | Pánve rotuje směrem dopředu pro přípravu na přistání                           | Aktivace stabilizačních svalů pánve pro udržení rovnováhy během rotace pánve                                 |
| Kyčel      | Kyčel se pohybuje vpřed s ohledem na zajištění plynulosti pohybu               | Aktivace [Iliopsoas](Iliopsoas) a [Rectus femoris](Rectus_femoris) pro zajištění správného pohybu kyčle      |
| Koleno     | Koleno se postupně natahuje, zajišťující plynulý přechod mezi fázemi          | Aktivace [Quadriceps](Quadriceps) pro stabilizaci kolene a přípravu na další kontakt s podložkou            |
| Ankle (kotník) | Kotník je stále v dorsiflexi, připravený pro kontakt s podložkou             | Aktivace [Tibialis anterior](Tibialis_anterior) pro kontrolu nad pohybem kotníku při přistání                |
