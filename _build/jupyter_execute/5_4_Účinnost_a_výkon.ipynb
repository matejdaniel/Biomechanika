# Účinnost přeměny energie

I když je **energie při přeměně zachována**, neznamená to, že se veškerá vstupní energie změní na **užitečnou práci nebo energii**. Ve skutečnosti je **výstup užitečné energie vždy menší** než vstupní energie, protože část energie se ztrácí (například ve formě tepla, tření nebo zvuku).

## Definice účinnosti

Účinnost procesu přeměny energie definujeme jako poměr **užitečné výstupní energie nebo práce** ku **celkové vstupní energii**:

$$
\eta = \frac{\text{užitečný výstup energie nebo práce}}{\text{celkový vstup energie}} = \frac{W_{\text{out}}}{E_{\text{in}}}
$$

# Výkon



> **Výkon** $P$ je fyzikální veličina, která udává, **jak rychle se vykonává práce** nebo **jak rychle se přenáší energie**.

$$P = \frac{W}{t}$$

- $P$ je výkon (např. ve wattech),
- $W$ je vykonaná práce nebo přenesená energie (např. v joulech),
- $t$ je čas, za který byla práce vykonána (v sekundách).

| Název jednotky       | Značka | Rovnost v SI jednotkách            | Běžné použití                      |
|----------------------|--------|------------------------------------|------------------------------------|
| Watt                 | W      | 1 W = 1 J/s                      | Elektřina, světelné zdroje         |
| Kilowatt             | kW     | 1 kW = 1000 W                    | Výkon spotřebičů, motorů           |
| Megawatt             | MW     | 1 MW = 1 000 000 W               | Elektrárny, průmyslové stroje      |
| Koňská síla (metrická) | hp     | 1 hp ≈ 735.5 W                   | Automobily, motory                 |
| Kalorie za sekundu   | cal/s  | 1 cal/s ≈ 4.184 W                | Výživové a biologické aplikace     |
| Erg za sekundu       | erg/s  | 1 erg/s = 1×10⁻⁷ W               | Fyzika (historicky)                |
| BTU za hodinu        | BTU/h  | 1 BTU/h ≈ 0.293 W                | Topení, klimatizace (USA)          |


> **Příkon** $P_0$ je fyzikální veličina, která udává, **kolik energie za jednotku času zařízení spotřebuje** (bez ohledu na to, jak efektivně ji přemění na užitečnou práci).

Zatímco **výkon** udává množství užitečné energie (např. mechanické práce nebo tepla) za sekundu,  
**příkon** udává, kolik **celkové energie** se příjme. 

$$
\text{Účinnost} = \frac{\text{Výkon}}{\text{Příkon}} \quad \Rightarrow \quad P = \eta \cdot P_0
$$

Stejně jako výkon se i příkon měří v jednotkách **watt (W)**.

| Jednotka    | Značka | Rovnost         |
|-------------|--------|-----------------|
| watt        | W      | 1 W = 1 J/s   |
| kilowatt    | kW     | 1 kW = 1000 W |
| megawatt    | MW     | 1 MW = 10⁶ W  |


## Shrnutí
- **Příkon = energie, která se přijímá.**
- **Výkon = energie, která se využije k užitečné práci.**
- Rozdíl mezi příkonem a výkonem určuje **účinnost** zařízení.
