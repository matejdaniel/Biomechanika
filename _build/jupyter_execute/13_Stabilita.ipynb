## Stabilita v biomechanice

- **Definice:** Stabilita je schopnost tlesa udret rovnovnou polohu nebo se do n vrtit po vychlen. V biomechanice se tk schopnosti lidskho tla udret nebo obnovit rovnovhu pi statickch (stn) nebo dynamickch (chze, bh) podmnkch.  

- **Typy stability:**  
  - **Statick stabilita:** Schopnost udret stabiln polohu bez pohybu (nap. rovnovn postoj).  
  - **Dynamick stabilita:** Schopnost udret rovnovhu bhem pohybu (nap. pi bhu nebo skoku).  

## Stabiln poloha

* **Stabiln rovnovn poloha**: pi vychlen se tit tlesa zdvh. Jakmile pestane psobit vychylujc sla, tit se vrt do pvodn polohy. Pi vychlen tlesa ze stl rovnovn polohy se zvyuje vka tit tlesa vzhledem k povrchu Zem a roste tak jeho potenciln thov energie. Tleso m ve stl rovnovn poloze tit v nejni mon poloze a jeho potenciln energie je nejmen.

## Kdy je lovk v statick rovnovze?

Pro dosaen rovnovhy mus bt splnny dv hlavn podmnky:  

1. **Translan rovnovha**  Souet vech sil psobcch na tlo mus bt roven nule.  
2. **Rotan rovnovha**  Souet vech moment sil psobcch na tlo vzhledem k jakmukoli bodu mus bt roven nule.  


### **1. Rovnovha sil (reakn sla a thov sla)**

- Na stojcho lovka psob dv hlavn sly:  
  - **Thov sla** (\( F_g \))  Smuje dol a psob v titi (**COG**), jej velikost je dna vztahem:  
    $$
    F_g = m \cdot g
    $$  
    kde \( m \) je hmotnost tla a \( g \) je gravitan zrychlen.  

  - **Reakn sla podloky** (\( R \))  Smuje vzhru a psob v mst kontaktu s podlokou (v bod **COP**  tit tlaku). Tato sla je reakc na thovou slu podle **tetho Newtonova zkona**.  

- **Podmnka pro rovnovhu sil:**  
  $$
  \sum F = 0
  $$  
  co znamen:  
  $$
  R = F_g
  $$

### **2. Rovnovha moment (otiv inky sil)**

- Aby lovk zstal v rovnovze, mus bt vyrovnny tak vechny momenty psobc na tlo.  
- **Moment sly** (\( M \)) je dn vztahem:  
  $$
  M = F \cdot d
  $$  
  kde \( F \) je sla a \( d \) je kolm vzdlenost od osy oten (nap. kotnkovho kloubu).  

- **Podmnka pro rotan rovnovhu:**  
  $$
  \sum M = 0
  $$  
  - Pokud se tit nachz pmo nad bodem kontaktu s podlokou (**COP**), nevznik dn moment, kter by tlo vychyloval.  
  - Pokud se tit vychl mimo zkladnu opory, vznik moment:  
    $$
    M = F_g \cdot d
    $$  
  Tlo mus aktivovat svaly a vytvoit kompenzan moment, aby zabrnilo pdu.  


| **Podmnka**                  | **Popis**                                                                                 | **Rovnice**                           |
|-------------------------------|-------------------------------------------------------------------------------------------|---------------------------------------|
| **Translan rovnovha**      | Souet vech sil psobcch na tlo mus bt roven nule.                                   | $\sum F = 0 $                       |
| **Vyrovnn thov sly**     | Reakn sla podloky mus bt rovna thov sle, ale opanho smru.                     | $R = F_g $                         |
| **Rotan rovnovha**         | Souet vech moment sil vzhledem k jakmukoli bodu mus bt roven nule.                   | $\sum M = 0$                       |
| **Stabiln poloha tit**   | Tnice (**LOG**) mus prochzet uvnit zkladny opory (**BOS**).                      |                                      |
| **Kompenzace otivho momentu** | Svaly mus vyvinout dostaten moment k vyrovnn ppadnho naruen rovnovhy.           | $M = F_g \cdot d $                  |

#### **Praktick pklad:**  
Pi **stoji na jedn noze** se tit posouv a thov ra se bl k hranici zkladny opory. Tlo aktivuje svaly dolnch konetin a trupu, aby vytvoilo kompenzan moment a udrelo rovnovhu. Pokud kompenzan moment nesta, dochz k pdu.  

## Zkladna opory

> Zkladna opory (**Base of Support - BOS**) oznauje plochu pod objektem nebo osobou, kter zahrnuje vechny body kontaktu s podprnou plochou. Tyto body kontaktu mohou bt sti tla, jako jsou **chodidla** nebo **ruce**, nebo tak pomcky, napklad **berle**, **hl**, nebo dokonce **idle**, na kter osoba sed.  

Zkladna opory je klovm faktorem ovlivujcm stabilitu tla. m vt a ir je zkladna opory, tm vt je schopnost tla udret rovnovhu. Stabilita je dosaena tehdy, kdy se **tnice (LOG)** nachz uvnit tto zkladny. Pokud thov ra peshne hranice zkladny opory, dochz k naruen rovnovhy a riziku pdu.

BOS se dynamicky mn v zvislosti na poloze tla a pouit externch oprnch pomcek. Napklad:  
- Pi **stn s nohama u sebe** je zkladna opory mal, co sniuje stabilitu.  
- Pi **rozkroenm postoji** se zkladna roziuje, co zvyuje stabilitu.  
- Pi **sedu na idli** zahrnuje zkladna opory kontaktn plochu mezi tlem a idl, stejn jako nohy dotkajc se podlahy.  
- Pokud lovk pouv **berle**, zkladna opory zahrnuje i body kontaktu mezi berlemi a zem.

- V **klinick praxi** se hodnot schopnost pacienta rozit svou zkladnu opory v reakci


## Stabilita a tit

Smr gravitan sly psobc na tlo smuje dol, k centru Zem a skrz tit. Tato linie gravitace je dleit pro pochopen a vizualizaci pi urovn schopnosti lovka udret rovnovhu. Kdy linie gravitace spadne mimo oprnou plochu (BOS), je nutn posturln reakce k udren rovnovhy.

Kdy linie gravitace spad do oprn plochy (BOS), objekt nebo osoba je stabiln. Kdy linie gravitace spadne mimo oprnou plochu (BOS), objekt nebo osoba je nestabiln.

Vzhledem k tomu, e linie gravitace mus spadat do oprn plochy (BOS), aby byly splnny kritria stability, je teba zvit nsledujc faktory:

* **Vt oprn plocha (BOS) zvyuje stabilitu** (linie gravitace se mus posunout na vt vzdlenost, aby spadla mimo oprnou plochu).
* **Ni tit (COG) zvyuje stabilitu** (je mn pravdpodobn, e linie gravitace spadne mimo oprnou plochu).

## **Posturln kyv (Postural Sway)**

> **Posturln kyv** oznauje mal, neustl pohyby tla, kter se objevuj pi pokusu udret stabiln vzpmenou polohu, napklad pi stn nebo sezen. Tyto vkyvy jsou pirozenou soust kontroly rovnovhy a jsou vsledkem jemnch prav polohy tla v reakci na vnitn i vnj podnty.

:width: 400px
:::

Lidsk tlo se pi pokusu o udren rovnovhy nikdy nepohybuje zcela staticky. I pi stn na mst dochz k neustlm drobnm pohybm, kter jsou zeny:  
* **Senzorickm systmem**
* Vstupy z **zrakovho systmu**, **vestibulrnho apartu** (rovnovn orgny ve vnitnm uchu) a **propriocepce** (vnmn polohy tla a napt ve svalech).  
* **Centrlnm nervovm systmem**  Vyhodnocuje smyslov informace a koordinuje odpov.  
* **Svalovou aktivitou**  Aktivace sval (nap. ltkovch a binch) upravuje polohu tla a zajiuje stabilitu.

### **Faktory ovlivujc posturln vkyv**  
| **Faktor**               | **Popis**                                                             |
|---------------------------|----------------------------------------------------------------------|
| **Zrak**                  | Zaven oi zvyuj rozsah posturlnho vkyvu.                      |
| **Pevnost zkladny opory**| U zkladna opory zvyuje vkyvy.                                 |
| **Vk**                   | S vkem se obvykle zvyuje rozsah posturlnho vkyvu.               |
| **nava**                 | Sniuje efektivitu svalov reakce, m zvyuje vkyv.               |
| **Nemoci**                | Neurologick onemocnn (nap. Parkinsonova choroba) mohou zvyovat nestabilitu. |

### **Men posturlnho vkyvu**  
Posturln vkyv se asto m pomoc **stabilometrie**  zazen sleduje pohyby bodu tit tlaku (**COP**). Vsledky se obvykle zobrazuj jako trajektorie pohybu na dvou osch:  
- **Anteroposteriorn smr** (pedozadn)    5-7 mm
- **Mediolaterln smr** (stranov)  3-4 mm 

### **Vznam v praxi**  
- V **klinick praxi** se posturln vkyv vyuv k diagnostice poruch rovnovhy.  
- V **rehabilitaci** pomh pi hodnocen pokroku u pacient po razech nebo s neurologickmi poruchami.  
- Ve **sportovn vd** se pouv k optimalizaci rovnovhy a prevence zrann u sportovc.
